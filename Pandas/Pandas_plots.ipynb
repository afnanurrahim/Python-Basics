```python
import pandas 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load the iris data from a given csv file into a data frame and print the shape of the data, type of the data and first and last 4 rows

```python
datafile=" Iris_data_sample.csv.csv"
```

```python
dataset=pandas.read_csv(datafile)
dataset
```

```python
print("SHAPE\n")
print(dataset.shape)
print("\nTYPE\n")
print(dataset.dtypes)
print("\nROWS\n")
print(dataset[:1])
print(dataset[-4:])
```

#### Replace the abnormal symbols i.e. ?? , ## to null value i.e. NaN

```python
dataset['PetalLengthCm']= dataset['PetalLengthCm'].replace("###",np.NaN)
dataset
```

#### Describe the dataset, and find number of observations, missing values and nan values

```python
print("Number of observations ",len(dataset.axes[0]))
```

```python
dataset.isnull().sum()
```

#### View basic statistical details like percentile, mean, std deviation, mode, variance, skewness and kutosis of iris data

```python
dataset.describe()
```

```python
print(dataset.mode())
print(dataset.var())
print(dataset.skew())
print(dataset.kurt())
```
