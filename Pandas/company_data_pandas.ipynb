```python
import pandas 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Read Total profit of all months and show it using a line plot

```python
data="company_sales_data.csv"
dataset=pandas.read_csv(data)
dataset
```

```python
plt.plot(dataset.month_number,dataset.total_profit)
plt.title("Profit data",color='red')
plt.xlabel("Month")
plt.ylabel("Total profit")
```

2. Get total profit of all months and show line plot with the following Style properties.
Generated line plot must include following Style properties: â€“

 Line Style dotted and Line-color should be red

 Show legend at the lower right location.

 X label name = Month Number

 Y label name = Sold units number

 Add a circle marker.

 Line marker color as read

 Line width should be 3

```python

plt.plot(dataset.month_number,dataset.total_profit,marker='o',color='red',linestyle='dotted',linewidth=3)
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(["red"],loc='lower right')
```

3. Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e.,
Separate Plotline for each product).

```python
plt.subplot(6, 1, 1)
plt.plot(dataset.month_number,dataset.facecream,label="facecream")

plt.subplot(6, 1, 2)
plt.plot(dataset.month_number,dataset.facewash,color='red',label="facewash")

plt.subplot(6, 1, 3)
plt.plot(dataset.month_number,dataset.toothpaste,color='black',label="toothpaste")

plt.subplot(6, 1, 4)
plt.plot(dataset.month_number,dataset.bathingsoap,color='blue',label="bathingsoap")

plt.subplot(6, 1, 5)
plt.plot(dataset.month_number,dataset.shampoo,color='green',label="shampoo")

plt.subplot(6, 1, 6)
plt.plot(dataset.month_number,dataset.moisturizer,color='yellow',label="moisturizer")
```
