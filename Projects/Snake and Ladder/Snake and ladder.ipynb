```python
import random
import time
import pygame
ladder={1:38,4:14,9:31,21:42,28:84,51:67,72:91,80:99}
snake={17:7,54:34,62:19,64:60,87:36,93:73,95:75,98:79}
x=0;y=0
def move(screen,pos,turn): #even if turn is not passed we can still use it as it is in the code where "move" is called
    dice=random.randint(1,6)
    string=f"Dice:{dice}"
    dice_display=font.render(string, True, green, blue)
    screen.blit(dice_display,(1100,400))
    pos=pos+dice
    if pos in snake:
        pos=snake[pos]
        SorL="Snake Bite"
    elif pos in ladder:
        pos=ladder[pos]
        SorL="Ladder"
    else:
        SorL=None
    if((int)(pos/10)%2==0): #move right (if even)
        if(pos%10==0):
            x=(((pos+1)%10)-1)*100+50
        else:
            x=((pos%10)-1)*100+50
        y=665-(int)(pos/10)*70
    else:                   # move left (if odd)
        if(pos%10==0):
            x=(10-((pos+1)%10))*100+50
        else:
            x=(10-(pos%10))*100+50
        y=665-(int)(pos/10)*70
    if SorL is not None:
        SorL_display=font.render(SorL, True, green, blue)
        screen.blit(SorL_display,(1100,450))
    return pos,x,y
```

```python
import pygame,sys
from pygame.locals import *
import io
try:
    # Python2
    from urllib2 import urlopen
except ImportError:
    # Python3
    from urllib.request import urlopen

pygame.init()
white = (255, 255, 255)
green = (0, 255, 0)
blue = (0, 0, 128)
    
imageUrl="https://5.imimg.com/data5/SELLER/Default/2021/2/HK/IO/BL/12304017/snakes-and-ladders-game-board-500x500.jpg"
imageStr=urlopen(imageUrl).read()
imageFile=io.BytesIO(imageStr)
display_surface = pygame.display.set_mode((1300,1000))
pygame.display.set_caption('Snake and Ladder')
Img = pygame.image.load(imageFile)
boardImg=pygame.transform.scale(Img,(1000,700))
display_surface.fill(blue)
display_surface.blit(boardImg, (0,0))

font = pygame.font.Font('freesansbold.ttf', 20)
player="PLAYER 1"
player_display= font.render(player, True, green, blue)
pos1=0
pos2=0
p1_vertex=[0,0]
p2_vertex=[0,0]
turn=1
dice=" "
dice_display=font.render(dice, True, green, blue)
snake_or_ladder=""
SorL_display=font.render(snake_or_ladder, True, green, blue)
winner=""
space="Enter space to roll the dice"
textSpace=font.render(space, True, green, blue)
while True:
    display_surface.fill(blue)
    display_surface.blit(boardImg, (0,0))
    pygame.draw.circle(display_surface, (0,0,0), (p1_vertex[0], p1_vertex[1]), 30, 30)
    p1_display= font.render(str(1), True, (255,255,255))
    display_surface.blit(p1_display,(p1_vertex[0], p1_vertex[1]))
    pygame.draw.circle(display_surface, (0,0,0), (p2_vertex[0], p2_vertex[1]), 30, 30)
    p2_display= font.render(str(2), True, (255,255,255))
    display_surface.blit(p2_display,(p2_vertex[0], p2_vertex[1]))
    if pos1>=100 :
        winner="PLAYER 1 WON THE GAME!!"
        textWin=font.render(winner, True, green, blue)
        display_surface.blit(textWin,(1000,250))
    elif pos2>=100:
        winner="PLAYER 2 WON THE GAME!!"
        textWin=font.render(winner, True, green, blue)
        display_surface.blit(textWin,(1000,250))
    else:
        display_surface.blit(player_display,(1100,250))
        
        display_surface.blit(textSpace,(1000,300))
        for event in pygame.event.get(KEYDOWN):
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    if turn==1:
                        player="PLAYER 2"
                        turn=2
                        pos1,p1_vertex[0],p1_vertex[1]=move(display_surface,pos1,turn)
                    else:
                        player="PLAYER 1"
                        turn=1
                        pos2,p2_vertex[0],p2_vertex[1]=move(display_surface,pos2,turn)
                    player_display= font.render(player, True, green, blue)
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
    pygame.display.update()
```
